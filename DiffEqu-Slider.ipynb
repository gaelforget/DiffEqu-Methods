# Creating an interactive plot

* For reference, the function file which implements this is titled `diffequ-slide.jl`

To begin, we initialize all the packages we'll need to create the interactive plot:

```julia
using Interact; 
using Colors;
using Plots;
```

Then, we create our interactive plot and widget using `@manipulate`:

```julia
C = randn(1)[1]
@manipulate for Time = 0:0.1:10
    plot([t -> sin(2*pi*t)/(2*pi)+C*t],0,Time, color=:red, label="Solution X(t)")
end
```

*Et voila!* We've made a nice interactive plot.
